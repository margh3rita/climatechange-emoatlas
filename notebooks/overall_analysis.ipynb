# Overall analysis

```python
import sys
sys.path.insert(0,'..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import heapq


ND_DIR = Path("../data/emoatlas_natural_democratics")
NR_DIR = Path("../data/emoatlas_natural_republicans")
SD_DIR = Path("../data/emoatlas_synthetic_democratics")
SR_DIR = Path("../data/emoatlas_synthetic_republicans")
```

Look at the error between zscores, democrats:

```python
for year in ['2016', '2017']:
    

    nd_df = pd.read_csv(ND_DIR / f"zscores_{year}.csv")

    sd_df = pd.read_csv(SD_DIR / f"zscores_{year}.csv")
    

    # compute the differences between the natural and synthetic z-scores for each emotion, as a dictionary
    diff_dems = dict(zip(nd_df['emotion'], (nd_df['zscore'] - sd_df['zscore']).abs()))

    # flop 3:
    flop_dems =   set(heapq.nlargest(3, diff_dems.values()))
    max_diff_dems_emotion = [emotion for emotion, diff in diff_dems.items() if diff in flop_dems]

    # top 3:
    top_dems =   set(heapq.nsmallest(3, diff_dems.values()))
    min_diff_dems_emotion = [emotion for emotion, diff in diff_dems.items() if diff in top_dems]

    print(f"Emotion with the largest difference in z-scores between natural and synthetic for democrats in {year}:", max_diff_dems_emotion)
    print(f"Absolute difference in z-scores between natural and syntheti:", flop_dems)
    print('\n')
    print(f"Emotion with the smallest difference in z-scores between natural and synthetic for democrats in {year}:", min_diff_dems_emotion)
    print(f"Absolute difference in z-scores between natural and syntheti:", top_dems)
    print('\n')


```

Look at the error between zscores, republicans:

```python
for year in ['2016', '2017']:
    

    nr_df = pd.read_csv(NR_DIR / f"zscores_{year}.csv")

    sr_df = pd.read_csv(SR_DIR / f"zscores_{year}.csv")
    

    # compute the differences between the natural and synthetic z-scores for each emotion, as a dictionary
    diff_reps = dict(zip(nr_df['emotion'], (nr_df['zscore'] - sr_df['zscore']).abs()))

    # flop 3:
    flop_reps =   set(heapq.nlargest(3, diff_reps.values()))
    max_diff_reps_emotion = [emotion for emotion, diff in diff_reps.items() if diff in flop_reps]

    # top 3:
    top_reps =   set(heapq.nsmallest(3, diff_reps.values()))
    min_diff_reps_emotion = [emotion for emotion, diff in diff_reps.items() if diff in top_reps]

    print(f"Emotion with the largest difference in z-scores between natural and synthetic for republicans in {year}:", max_diff_reps_emotion)
    print(f"Absolute difference in z-scores between natural and synthetic:", flop_reps)
    print('\n')
    print(f"Emotion with the smallest difference in z-scores between natural and synthetic for republicans in {year}:", min_diff_reps_emotion)
    print(f"Absolute difference in z-scores between natural and synthetic:", top_reps)
    print('\n')
```
